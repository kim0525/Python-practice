```python
img_url = "https://www.reshot.com/free-stock-photos/search/animal/"
html = requests.get(img_url).text
soup = BeautifulSoup(html, "html.parser")
```

```python
soup
```

```python
div_animals = soup.find_all("div", {"class":"item-grid-item"})
div_animals[1].find("img")
```

```python
div_animals[1].find("img").get("src")
```

```python
len(div_animals)
```

```python
animal_url = []
for i in range(1, len(div_animals)):
    animal_url.append(div_animals[i].find("img").get("src"))
    print(animal_url[i-1])
```

```python
for i in range(1, 10):
    
    file_name = os.path.basename(animal_url[i])
    html_image = requests.get(animal_url[i])

    fp = open(file_name, "wb")
    chunk_size = 1000000
    for chunk in html_image.iter_content():
        fp.write(chunk)
    fp.close()
    
```
